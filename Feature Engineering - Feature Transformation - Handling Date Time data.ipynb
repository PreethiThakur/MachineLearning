```python
import pandas as pd
import numpy as np
```

```python
orders = pd.read_csv('./Datasets/orders.csv', usecols=['Order Number', 'Order Date', 'Quantity'])
orders
```

```python
orders.info()
```

```python
# parsing to datetime
orders['Order Date'] = pd.to_datetime(orders['Order Date'])
orders.dtypes
```

```python
# extract year
orders['order_year'] = orders['Order Date'].dt.year

# extract day
orders['order_day'] = orders['Order Date'].dt.day  

# day_name
orders['order_dayname'] = orders['Order Date'].dt.day_name() 

# extract month
orders['order_month'] = orders['Order Date'].dt.month_name()

# dayofweek
orders['order_dayofweek'] = orders['Order Date'].dt.dayofweek

orders
```

```python
# extracting if there are saturday and sunday orders
orders['Weekend'] = np.where(orders['order_dayname'].isin(['Saturday', 'Sunday']), 1,0)
```
